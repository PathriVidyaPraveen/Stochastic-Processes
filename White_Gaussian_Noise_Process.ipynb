A common model for physical noise, such as resistor noise due to electron motion fluctuations
in an electric field, is termed white Gaussian noise (WGN). It is assumed that the noise has
been sampled in time to yield a discrete-time continuous-valued random process {Xn : n ∈ N}.
The WGN random process is defined to be the IID process where Xn ∼ N(0, σ2), n ∈ N.
Generate an overlaid scatterplot (matplotlib.pyplot.scatter) of 50 realizations of the WGN
process with variance 1.

```python
import numpy as np
import matplotlib.pyplot as plt

num_realizations = 50
num_samples = 50
mean = 0
variance = 1

# generate WGN samples

X = np.random.normal(mean,np.sqrt(variance),size=(num_realizations,num_samples))

plt.figure(figsize=(12,6))

for i in range(num_realizations):
  x = np.arange(1,num_samples+1)
  y = X[i,:]
  plt.scatter(x,y,s=10,alpha=0.6)

plt.title('50 realizations of WGN with unit variance')
plt.xlabel('Time index n')
plt.ylabel('Xn ~ N(0,1)')
plt.grid(False)
plt.show()
```
