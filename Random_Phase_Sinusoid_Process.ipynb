Consider the random-phase sinusoid process {Xn = cos (2π ∗ 0.1n + Θ), n ∈ N}, where Θ ∼
unif[0, 2π]. Generate and plot on the same figure 5 sample paths of the process for n ∈
{1, 2, · · · , 50}.
Remark: This random process is frequently used to model an analog sinusoid whose phase is
unknown and that has been sampled by an analog-to-digital converter.



```python
import numpy as np
import matplotlib.pyplot as plt

pi = np.pi

thetas = np.random.uniform(0,2*pi,5)
colors = ['blue','red','green','purple','magenta']
i=0
n = np.arange(1,51)
plt.figure(figsize=(12,6))
for theta in thetas:
  y = np.cos(2*pi*0.1*n + theta)
  x = n
  plt.plot(x,y,color=colors[i],label=f"Path {i+1}")
  i+=1

plt.xlabel('n')
plt.ylabel('Xn = cos(2pi*0.1*n + theta)')
plt.title('5 Sample Paths of Random Phase Sinusoid Process')
plt.grid(False)
plt.legend()
plt.show()
```
